```python
# Yearly returns for individual companies
ind_er = selected_stocks.resample('Y').last().pct_change().mean() 
ind_er
```

```python
# Portfolio returns
w = [1/len(selected_stocks.columns)]*len(selected_stocks.columns)
port_er = (w*ind_er).sum()
port_er
```

```python
# Volatility is given by the annual standard deviation. We multiply by 252 because there are 252 trading days/year.
ann_sd = selected_stocks.pct_change().apply(lambda x: np.log(1+x)).std().apply(lambda x: x*np.sqrt(252)) 
ann_sd
```

```python
assets = pd.concat([ind_er,ann_sd], axis=1)
assets.columns = ['Returns', 'Volatility']
assets
```

The graph of the efficient frontier is plotted by running a loop 10000 times, with each iteraction considering different weights for each index fund and calculating the return and volatility of the specific portfolio combination

```python
p_ret = [] # Define an empty array for portfolio returns 
p_vol = [] # Define an empty array for portfolio volatility 
p_weights = [] # Define an empty array for asset weights
num_assets = len(selected_stocks.columns) 
num_portfolios = 10000
```

```python
for portfolio in range(num_portfolios): 
    weights = np.random.random(num_assets) 
    weights = weights/np.sum(weights) 
    p_weights.append(weights)
    returns = np.dot(weights, ind_er) # Returns are the product of individual expected returns of asset and its # weights
    p_ret.append(returns)
    var = cov_matrix.mul(weights, axis=0).mul(weights, axis=1).sum().sum() # Portfolio Variance 
    sd = np.sqrt(var) #Daily standard deviation
    ann_sd = sd*np.sqrt(252) #Annual standard deviation = volatility 
    p_vol.append(ann_sd)
```

```python
data = {'Returns':p_ret, 'Volatility':p_vol}
for counter, symbol in enumerate(selected_stocks.columns.tolist()): 
    #print(counter, symbol)
    data[symbol+' weight'] = [w[counter] for w in p_weights]
portfolios = pd.DataFrame(data)
portfolios.head() # Dataframe of the 10000 portfolios created
```

```python
 
# Plot efficient frontier
portfolios.plot.scatter(x='Volatility', y='Returns', marker='o', s=10, alpha=0.3, grid=True, figsize=[5,5])
```

```python
min_vol_port = portfolios.iloc[portfolios['Volatility'].idxmin()] 
# idxmin() gives us the minimum value in the column specified. 
min_vol_port
```

```python
# plotting the minimum volatility portfolio
plt.subplots(figsize=[6,5])
plt.scatter(portfolios['Volatility'], portfolios['Returns'],marker='o', s=10, alpha=0.3) 
plt.scatter(min_vol_port[1], min_vol_port[0], color='r', marker='*', s=200)
```

Sharpe ratio which is a measure of the average return earned in excess of the risk-free rate per unit of risk is used to select the optimal risky portfolio. The portfolio with the highest sharpe ratio is the most optimal.

The risk-free rate is the return on the investment with zero risk. Here we use the 3-month treasury bill yield as the risk-free rate which is 0.35% (sourced from Bloomberg, 2022).

```python
# Finding the optimal portfolio
rf = 0.0035 # risk factor
optimal_risky_port = portfolios.iloc[((portfolios['Returns']-rf)/portfolios['Volatility']).idxmax()] 
optimal_risky_port
```

```python
# Plotting optimal portfolio
plt.subplots(figsize=(5, 6))
plt.scatter(portfolios['Volatility'], portfolios['Returns'],marker='o', s=10, alpha=0.3) 
plt.scatter(min_vol_port[1], min_vol_port[0], color='r', marker='*', s=500) 
plt.scatter(optimal_risky_port[1], optimal_risky_port[0], color='g', marker='*', s=500)
```
